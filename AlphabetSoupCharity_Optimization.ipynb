## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
df = application_df.drop(columns=['EIN'])
df
```

```python
threshold = 300

# Get the frequency of each name
name_counts = df['NAME'].value_counts()

# Replace rare names with "Other"
df['NAME_BINNED'] = df['NAME'].apply(
    lambda x: x if name_counts[x] > threshold else 'Other'
)

# Drop the original NAME column
df = df.drop(columns=['NAME'])

# Check the result
print(df['NAME_BINNED'].value_counts())
```

```python
# Determine the number of unique values in each column.
df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = ['T25', 'T14', 'T29', 'T15', 'T17', 'T2', 'T12']

# Replace column names
for app in application_types_to_replace:
    df['APPLICATION_TYPE'] = df['APPLICATION_TYPE'].replace(app,"Other")

# make sure replacement was successful
df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
df['CLASSIFICATION'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
class_counts = df['CLASSIFICATION'].value_counts()

# Filter and show only values greater than 1
class_counts_filtered = class_counts[class_counts > 1]

# Display the result
print(class_counts_filtered)
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
cutoff = 1000

class_counts = application_df['CLASSIFICATION'].value_counts()

classifications_to_replace = class_counts[class_counts < cutoff].index.tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    df['CLASSIFICATION'] = df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
df['CLASSIFICATION'].value_counts()
```

```python
df['ASK_AMT'].value_counts()
```

```python
cutoff = 4

amt_counts = df['ASK_AMT'].value_counts()

amt_to_replace = amt_counts[amt_counts < cutoff].index.tolist()

# Replace in dataframe
for cls in amt_to_replace:
    df['ASK_AMT'] = df['ASK_AMT'].replace(cls,"Other")

# Check to make sure replacement was successful
df['ASK_AMT'].value_counts()
```

```python
df.nunique()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
df = pd.get_dummies(df)
df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = df["IS_SUCCESSFUL"]
X = df.drop(columns='IS_SUCCESSFUL')

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net
input_features = X_train.shape[1]

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=64, activation='relu', input_dim=input_features))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=32, activation='relu'))

# Third hidden layer (slightly smaller to force abstraction)
nn.add(tf.keras.layers.Dense(units=16, activation='relu'))

# Add dropout for regularization (reduce dropout rate to retain more neurons)
nn.add(tf.keras.layers.Dropout(0.1))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Compile the model with Adam optimizer (reduced learning rate)
optimizer = tf.keras.optimizers.Adam(learning_rate=0.0005)
nn.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics= ['accuracy'])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs= 100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
